### Replacements
Gender: {'Male': 0, 'Female': 1}
Customer Type: {'disloyal Customer': 0, 'Loyal Customer': 1}
Type of Travel: {'Personal Travel': 0, 'Business travel': 1}
Class: {'Eco': 0, 'Eco Plus': 1, 'Business': 2}
satisfaction: {'neutral or dissatisfied': 0, 'satisfied': 1}

```python
#df_r is 'with replacements'
df_r = df.replace({'Gender': 'Male', 'Customer Type': 'disloyal Customer', 'Type of Travel': 'Personal Travel',
           'Class': 'Eco', 'satisfaction': 'neutral or dissatisfied'}, 0)
df_r = df_r.replace({'Gender': 'Female', 'Customer Type': 'Loyal Customer', 'Type of Travel': 'Business travel',
                    'Class': 'Eco Plus', 'satisfaction': 'satisfied'}, 1)
df_r = df_r.replace({'Class': 'Business'}, 2)
df_r
```

```python
# preparing data for split
df_r_columns = list(df_r.columns.values)
x_columns = df_r_columns[:-1]
y_column = df_r_columns[-1]
x = df_r[x_columns]
y = df_r[y_column]
```

```python
# 70% for training, 30% for testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
clf = RandomForestClassifier(max_depth=2, random_state=0)
clf = clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
```

```python
print("Accuracy:", metrics.accuracy_score(y_test, y_pred)) # accuracy is nice
# perhaps stacking / blending can make it better
```

```python
# trying another depth
clf_2 = RandomForestClassifier(max_depth=5, random_state=0)
clf_2 = clf_2.fit(x_train, y_train)
y_pred = clf_2.predict(x_test)
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

The bigger the depth, the more classifier is prone to overfitting (with depth 5 it is probably overfitted)

```python
# trying another depth
clf_3 = RandomForestClassifier(max_depth=3, random_state=0)
clf_3 = clf_3.fit(x_train, y_train)
y_pred = clf_3.predict(x_test)
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

Depth 3 seems to be promising
